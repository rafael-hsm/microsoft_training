# Training for Regression - Supervised Machine Learning
## In this notebook, we'll focus on regression, using an example based on a real study in which data for a bicycle sharing scheme was colected and used to predic the number of rentals based on seasonality and weather conditions.

```python
#%pip install pandas
#%pip install matplotlib
#%pip install scikit-learn
```

```python
import pandas as pd

# load the training dataset
url_file = 'https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/daily-bike-share.csv'

bike_data = pd.read_csv(url_file)
bike_data.shape
bike_data.head()
```

```python
# Let's add a new column named day to the dataframe by extracting the day component
bike_data['day'] = pd.DatetimeIndex(bike_data['dteday']).day
bike_data.head(32)
```

```python
# Analysis of the data by examining a few key descriptive statistics.
numeric_features = ['temp', 'atemp', 'hum', 'windspeed']
bike_data[numeric_features + ['rentals']].describe()
```

```python
import matplotlib.pyplot as plt


# This ensures plots are displayed inline in the Jupyter notebook
%matplotlib inline

# Get the label column
label = bike_data['rentals']

# Create a figure for 2 subplots (2rows, 1 column)
fig, ax = plt.subplots(2, 1, figsize=(9,12))

# Plot the histogram
ax[0].hist(label, bins=100)
ax[0].set_label('Frequency')

# Add lines for the mean, median, and mode
ax[0].axvline(label.mean(), color='magenta', linestyle='dashed', linewidth=2)
ax[0].axvline(label.median(), color='cyan', linestyle='dashed', linewidth=2)

# Plot the boxplot
ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

# Add a title to the Figure
fig.suptitle('Rental Distribution')

# Show the figure
fig.show()
```

```python
# Plot a histogram for each numeric feature
for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = bike_data[col]
    feature.hist(bins=100, ax=ax)
    ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
```

```python
import numpy as np

# plot a bar plot for each categorical feature count
categorical_features = ['season', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit', 'day']

for col in categorical_features:
    counts = bike_data[col].value_counts().sort_index()
    fig = plt.figure(figsize=(9,6))
    ax = fig.gca()
    counts.plot.bar(ax=ax, color='steelblue')
    ax.set_title(col+ 'counts')
    ax.set_xlabel(col)
    ax.set_ylabel("Frequency")
    
plt.show()
    
    
```

```python
# Correlation statistic to quantify the apparent relationship.
for col in numeric_features:
    fig = plt.figure(figsize=(9,6))
    ax = fig.gca()
    feature = bike_data[col]
    label = bike_data['rentals']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs ' + col + '- correlation: ' + str(correlation))
    
plt.show()
```

```python
# plot a boxplot for the label by each categorical feature
for col in categorical_features:
    fig = plt.figure(figsize=(9,6))
    ax = fig.gca()
    bike_data.boxplot(column='rentals', by=col, ax=ax)
    ax.set_title('Label by ' + col)
    ax.set_ylabel("Bike Rentals")
    
plt.show()
```

```python
# Train a Regression Model
# Separate features and labels
X, y = bike_data[['season', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed']].values, bike_data['rentals'].values
print('Features: ', X[:10], 'n\Labels: ', y[:10], sep='\n')
```

```python
from sklearn.model_selection import train_test_split

# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print(f'Training Set: {X_train.shape[0]} rows\nTest Set: {X_test.shape[0]}')
```

```python
# Train the model
from sklearn.linear_model import LinearRegression

# Fit a linear regression model on the training set
model = LinearRegression().fit(X_train, y_train)
print(model)
```

```python
# Evaluate the Trained Model
import numpy as np

predictions_ = model.predict(X_test)
np.set_printoptions(suppress=True)
print('Predicted labels: ', np.round(predictions_)[:10])
print('Actual labels  : ', y_test[:10])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(y_test, predictions_)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')

# overlay the regression line
z = np.polyfit(y_test, predictions_, 1)
p = np.poly1d(z)
plt.plot(y_test, p(y_test), color='magenta')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions_)
print(f"MSE: {mse}")

rmse = np.sqrt(mse)
print(f"RMSE: {rmse}")

r2 = r2_score(y_test, predictions_)
print(f"R2: {r2}")
```
